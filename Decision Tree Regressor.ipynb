# Decision Tree Classifier

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
ps=pd.read_csv('Position_Salaries.csv')
ps
```

```python
ps.shape
```

```python
#select input ansd output
X=ps[["Level"]] #input (always2D)
Y=ps["Salary"] #traget (1D)
```

```python
#check null values
ps.isnull().sum()
```

```python
ps.dtypes
```

```python
#apply Decision TreeRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
#create object of DecisionTreeRegressor class
dtr=DecisionTreeRegressor(criterion='mse',random_state=1)
```

```python
#train the model we r not using train_test_split because the dataste is to small we use train_test_split where the dataset
#is atleast 100 or more then 100
dtr.fit(X,Y)
```

```python
#testing tye model 
Y_pred=dtr.predict([[6.5]])
```
